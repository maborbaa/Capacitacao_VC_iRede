````python
import numpy as np
import matplotlib.pyplot as plt
# Vamos usar a imagem clássica da astronauta que já vem no scikit-image
from skimage.data import astronaut ## pode buscar outras imagens
````

---
### Prática 2: Isolamento de Canais RGB
**Objetivo:** Decompor a matriz 3D da imagem em 3 matrizes 2D independentes, analisando a intensidade luminosa de cada espectro cromático.

````python
#importar a imagem
imagem = astronaut()

# Isolando os canais
_r = imagem[:,:,0]
_g = imagem[:,:,1]
_b = imagem[:,:,2]

#utilizar subplots para exibir tudo lado a lado
fig, axes = plt.subplots(1,4,figsize=(15,5)) #subplot(int,int,index)

# Quadro 0: Original
axes[0].imshow(imagem)
axes[0].set_title("Original (RGB)")
axes[0].axis('off')

# Quadro 1: Canal R
axes[1].imshow(_r, cmap='Reds')
axes[1].set_title("Canal R")
axes[1].axis('off')

# Quadro 2: Canal G
axes[2].imshow(_g, cmap='Greens')
axes[2].set_title("Canal G")
axes[2].axis('off')

# Quadro 3: Canal B
axes[3].imshow(_b, cmap='Blues')
axes[3].set_title("Canal B")
axes[3].axis('off')

plt.tight_layout()
plt.show()
````

---
### Prática 3: Escala de Cinza e Espaço HSV
**Objetivo:** Reduzir o custo computacional com a Escala de Cinza e isolar o brilho da cromaticidade usando o espaço de cor HSV (Matiz, Saturação, Valor).

````python
from skimage.color import rgb2gray, rgb2hsv

# 1. Convertendo para Escala de Cinza (1 canal)
imagem_cinza = rgb2gray(imagem)

# 2. Convertendo para HSV e isolando os canais
imagem_hsv = rgb2hsv(imagem)

canal_h = imagem_hsv[:, :, 0] # H (Hue / Matiz) -> O tipo da cor
canal_s = imagem_hsv[:, :, 1] # S (Saturation / Saturação) -> A pureza da cor
canal_v = imagem_hsv[:, :, 2] # V (Value / Brilho) -> A intensidade da luz

# 3. Exibindo os resultados
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(imagem_cinza, cmap='gray')
axes[0].set_title("Escala de Cinza (1 Canal)")
axes[0].axis('off')

# Usamos cmap='hsv' para visualizar a roda de cores real no Matiz
axes[1].imshow(canal_h, cmap='hsv')
axes[1].set_title("Canal H (Matiz)")
axes[1].axis('off')

axes[2].imshow(canal_s, cmap='gray')
axes[2].set_title("Canal S (Saturação)")
axes[2].axis('off')

axes[3].imshow(canal_v, cmap='gray')
axes[3].set_title("Canal V (Brilho)")
axes[3].axis('off')

plt.tight_layout()
plt.show()
````
